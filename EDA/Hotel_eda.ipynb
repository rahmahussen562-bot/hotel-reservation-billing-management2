```python
#Define categorical columns for analysis
categorical_columns = categorical_columns.drop(['name','email'], errors='ignore') 
#name is excluded as it contains unique values (High Cardinality)to avoid treating every unique person as a 'rare category
# Loop through each column to identify rare categories
for col in categorical_columns:
    # Calculate the frequency as a percentage
    relative_freq = df_new[col].value_counts(normalize=True) * 100

    # Filter categories that appear in less than 3%
    rare_categories = relative_freq[relative_freq < 3]

    if not rare_categories.empty:
        print(f"Column: [{col}]")
        for category, percentage in rare_categories.items():
            print(f" - Category '{category}': {percentage:.2f}%")
       
        df_new[col] = df_new[col].replace(rare_categories.index, 'Other')
        
        print(f" >> Decision: Grouped into 'Other' successfully.")
        print("-" * 30)
    else:
        print(f"Column: [{col}] - No rare categories found.")
        print("-" * 30)
```
